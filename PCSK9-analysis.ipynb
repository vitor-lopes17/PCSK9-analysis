```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
sns.set(style="whitegrid")
```

```python
#RMSD of PCSK9 (0 - 1200 ns)

PCSK91 = pd.read_fwf("C:\\Users\\T-Gamer\\Documents\\IC\\2020\\mdanalysis\\RMSD-Full.xvg")




plt.ylim(0.1, 0.35)
plt.rcParams['axes.labelsize'] = 25
plt.rcParams['axes.titlesize'] = 20
plt.title('Apo-PCSK9 RMSD in 1200 ns of MD')
sns.lineplot(x='Time(ns)', y='RMSD', data=PCSK91, color = "grey")
sns.set(style="ticks", rc={"lines.linewidth": 0.5})



#plt.savefig('RMSD-full.pdf', dpi=600)
#plt.show()

```
